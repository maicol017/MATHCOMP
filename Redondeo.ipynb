```python
import math

def taylor_series_exp_neg_x(x, tol=1e-8):
    term = 1  # First term of the series
    sum_series = term
    n = 1
    while abs(term) > tol:
        term *= -x / n
        sum_series += term
        n += 1
    return sum_series, n-1

x = 0.85
result, iterations = taylor_series_exp_neg_x(x)
print(f"Resultado: {result:.8f}, Iteraciones: {iterations}")
```
