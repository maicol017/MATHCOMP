```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Datos
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 6.6, 8.5, 10.8, 14.2])

# Transformación logarítmica
log_y = np.log(y)

# Regresión lineal
slope, intercept = np.polyfit(x, log_y, 1)

# Ecuación de la recta
y_pred = np.exp(intercept) * np.exp(slope * x)

# Gráfico
plt.plot(x, y, 'o', label='Datos originales')
plt.plot(x, y_pred, '-', label='Regresión lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste logarítmico')
plt.legend()
plt.show()

# Ecuación de la potencia
print(f'y = {np.exp(intercept):.2f} * x^{slope:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([1, 2, 3, 4, 5])
y = np.array([3.5, 5.5, 6.8, 7.8, 8.3])

# Calcula la regresión lineal
slope, intercept = np.polyfit(x, y, 1)

# Ecuación de la recta
equation = f"y = {slope:.2f}x + {intercept:.2f}"

# Gráfico de los datos y la recta de regresión
plt.plot(x, y, 'o', label='Datos')
plt.plot(x, slope * x + intercept, '-', label='Regresión lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Regresión lineal: {equation}')
plt.legend()
plt.show()

# Imprime la ecuación de la recta
print(f"Ecuación de la recta de regresión: {equation}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([1, 3, 5, 7, 9, 11, 13, 15])
y = np.array([2.1, 3.2, 3.8, 4, 4.2, 4.4, 4.5, 4.7])

# Modelos
# -------------------------------------------------------------------------------------------------------------

# Modelo lineal
coef_lineal = np.polyfit(x, y, 1)
y_lineal = coef_lineal[0] * x + coef_lineal[1]

# Modelo cuadrático
coef_cuadratico = np.polyfit(x, y, 2)
y_cuadratico = coef_cuadratico[0] * x**2 + coef_cuadratico[1] * x + coef_cuadratico[2]

# Modelo exponencial
coef_exponencial = np.polyfit(np.log(x), y, 1)
y_exponencial = np.exp(coef_exponencial[1]) * np.exp(coef_exponencial[0] * np.log(x))

# Modelo de razón de crecimiento
coef_razon = np.polyfit(x, np.log(y), 1)
y_razon = np.exp(coef_razon[1]) * np.exp(coef_razon[0] * x)

# Gráfico
# -------------------------------------------------------------------------------------------------------------

plt.figure(figsize=(10, 6))
plt.scatter(x, y, label="Datos")
plt.plot(x, y_lineal, label="Lineal")
plt.plot(x, y_cuadratico, label="Cuadrático")
plt.plot(x, y_exponencial, label="Exponencial")
plt.plot(x, y_razon, label="Razón de Crecimiento")

plt.title("Regresión de Datos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
