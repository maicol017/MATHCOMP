```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
data = pd.read_csv('SaratogaHouses.csv')
x = data['lotSize']
y = data['price']

# Transformacin logartmica
log_y = np.log(y)

# Regresin
pendiente, intercepcin = np.polyfit(x, log_y, 1)

# Ecuacin de la recta
y_pred = np.exp(intercepcin) * np.exp(pendiente * x)

# Grfico
plt.plot(x, y, 'o', label='Datos originales')
plt.plot(x, y_pred, '-', label='Regresin lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste logartmico')
plt.legend()
plt.show()

# Ecuacin de la potencia
print(f'y = {np.exp(intercepcin):.2f} * x^{pendiente:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 6.6, 8.5, 10.8, 14.2])

# Modelo lineal
coef_lineal = np.polyfit(x, y, 1)
y_lineal = coef_lineal[0] * x + coef_lineal[1]

# Modelo cuadrtico
coef_cuadratico = np.polyfit(x, y, 2)
y_cuadratico = coef_cuadratico[0] * x**2 + coef_cuadratico[1] * x + coef_cuadratico[2]

# Modelo exponencial
coef_exponencial = np.polyfit(np.log(x), y, 1)
y_exponencial = np.exp(coef_exponencial[1]) * np.exp(coef_exponencial[0] * np.log(x))

# Modelo de razn de crecimiento
coef_razon = np.polyfit(x, np.log(y), 1)
y_razon = np.exp(coef_razon[1]) * np.exp(coef_razon[0] * x)

# Grfico
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label="Datos")
plt.plot(x, y_lineal, label="Lineal")
plt.plot(x, y_cuadratico, label="Cuadrtico")
plt.plot(x, y_exponencial, label="Exponencial")
plt.plot(x, y_razon, label="Razn de Crecimiento")
plt.title("Regresin de Datos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
