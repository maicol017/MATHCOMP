# Tamaños de bits que existen para las palabras

Unidades de Almacenamiento

Bit (b): La unidad más pequeña de información.

Byte (B): Equivale a 8 bits.

Kilobyte (KB): 1 KB = 1024 bytes.

Megabyte (MB): 1 MB = 1024 KB.

Gigabyte (GB): 1 GB = 1024 MB.

Terabyte (TB): 1 TB = 1024 GB.

Petabyte (PB): 1 PB = 1024 TB.

Exabyte (EB): 1 EB = 1024 PB.

Zettabyte (ZB): 1 ZB = 1024 EB.

Yottabyte (YB): 1 YB = 1024 ZB.

Unidades de Velocidad de Transferencia
Bits por segundo (bps): Unidad básica de velocidad de transferencia.
Kilobits por segundo (Kbps): 1 Kbps = 1000 bps.

Megabits por segundo (Mbps): 1 Mbps = 1000 Kbps.

Gigabits por segundo (Gbps): 1 Gbps = 1000 Mbps.

Terabits por segundo (Tbps): 1 Tbps = 1000 Gbps.

Unidades de Frecuencia

Hercio (Hz): Unidad básica de frecuencia.

Kilohertz (kHz): 1 kHz = 1000 Hz.

Megahertz (MHz): 1 MHz = 1000 kHz.

Gigahertz (GHz): 1 GHz = 1000 MHz.
