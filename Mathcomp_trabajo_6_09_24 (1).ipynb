```python
# Importamos las librerías necesarias
import math
import itertools

# 1. ¿Cuántos números entre 1 y 100 (inclusive) son divisibles entre 5 o 8?
def divisible_5_8():
    count = 0
    for i in range(1, 101):
        if i % 5 == 0 or i % 8 == 0:
            count += 1
    return count

# 2. ¿Cuántos equipos diferentes de tres personas pueden originarse si se tienen cinco personas para elegir entre ellas?
def equipos_tres_personas():
    return math.comb(5, 3)

# 3. ¿Cuántas permutaciones de tres elementos se forman con tres objetos?
def permutaciones_tres_elementos():
    return math.perm(3, 3)

# 4. ¿En cuántas formas diferentes pueden sacarse cuatro cartas (a la vez) de un paquete de 52 cartas?
def combinaciones_cuatro_cartas():
    return math.comb(52, 4)

# 5. ¿Cuántas señas diferentes, cada una de 8 banderas colocadas en línea vertical, pueden formarse con 4 banderas rojas, 3 blancas y una azul?
def permutaciones_banderas():
    return math.factorial(8) // (math.factorial(4) * math.factorial(3) * math.factorial(1))

# 6. Cuántos libros distintos de matemáticas, seis diferentes de física y dos diferentes de química se colocan en un estante
# a) ¿Los libros de cada asignatura deben estar todos juntos?
def libros_juntos():
    return math.factorial(3) * math.factorial(5) * math.factorial(6) * math.factorial(2)

# b) ¿Solamente los libros de matemáticas deben estar juntos?
def libros_matematicas_juntos():
    return math.factorial(8) * math.factorial(5)

# 7. Calcula el número de formas en que un ejecutivo puede elegir de 3 a 15 empleados para un ascenso.
def elegir_empleados():
    total = 0
    for i in range(3, 16):
        total += math.comb(15, i)
    return total

# 8. Calcula el número de formas en que un capataz puede escoger a 12 de 18 trabajadores para asignarles trabajo en tiempo extra.
def escoger_trabajadores():
    return math.comb(18, 12)

# 9. ¿Cuántos comités de 4 personas se pueden formar con 9 personas?
def comites_cuatro_personas():
    return math.comb(9, 4)

# 10. ¿De cuántas maneras se puede escoger un comité compuesto por 3 hombres y 2 mujeres, de un grupo de 7 hombres y 5 mujeres?
def comite_hombres_mujeres():
    return math.comb(7, 3) * math.comb(5, 2)

# 11. De un total de 5 technoarquitectos y 7 arquitecto mágico, se forma un comité de 2 technoarquitectos y 3 arquitectos mágicos.
# a. Puede formarlo cualquier tecnoarquitecto y cualquier arquitecto mágico.
def comite_tecnoarquitectos_arqmagicos():
    return math.comb(5, 2) * math.comb(7, 3)

# b. Un arquitecto mágico determinado debe pertenecer al comité.
def comite_arqmagico_determinado():
    return math.comb(5, 2) * math.comb(6, 2)

# c. Dos technoarquitectos determinados no deben estar en el comité.
def comite_sin_tecnoarquitectos():
    return math.comb(3, 2) * math.comb(7, 3)

# Ejecución de las funciones
print("1. Números divisibles entre 5 o 8:", divisible_5_8())
print("2. Equipos de tres personas:", equipos_tres_personas())
print("3. Permutaciones de tres elementos:", permutaciones_tres_elementos())
print("4. Combinaciones de cuatro cartas:", combinaciones_cuatro_cartas())
print("5. Permutaciones de banderas:", permutaciones_banderas())
print("6a. Libros de cada asignatura juntos:", libros_juntos())
print("6b. Libros de matemáticas juntos:", libros_matematicas_juntos())
print("7. Formas de elegir empleados:", elegir_empleados())
print("8. Formas de escoger trabajadores:", escoger_trabajadores())
print("9. Comités de cuatro personas:", comites_cuatro_personas())
print("10. Comité de 3 hombres y 2 mujeres:", comite_hombres_mujeres())
print("11a. Comité de tecnoarquitectos y arquitectos mágicos:", comite_tecnoarquitectos_arqmagicos())
print("11b. Comité con un arquitecto mágico determinado:", comite_arqmagico_determinado())
print("11c. Comité sin dos tecnoarquitectos determinados:", comite_sin_tecnoarquitectos())

# 12. Con 7 consonantes y 5 vocales diferentes, ¿cuántas palabras pueden formarse que consten de 4 consonantes y 3 vocales?
def palabras_consonantes_vocales():
    consonantes = math.comb(7, 4)
    vocales = math.comb(5, 3)
    return consonantes * vocales * math.factorial(7)

# 13. ¿Cuántas permutaciones se pueden hacer con un grupo de 40 alumnos tomando equipos de 4 personas? Y cuántas combinaciones?
def permutaciones_combinaciones_alumnos():
    permutaciones = math.perm(40, 4)
    combinaciones = math.comb(40, 4)
    return permutaciones, combinaciones

# 14. ¿De cuántas maneras se puede ordenar 3 libros en un librero si se escogen 2 libros a la vez?
def ordenar_libros():
    return math.perm(3, 2)

# 15. Un disco compacto puede ser comprado en cualquiera de las 5 tiendas. ¿De cuántas maneras se puede seleccionar 3 de las 5 tiendas?
def seleccionar_tiendas():
    return math.comb(5, 3)

# 16. Determine el número de formas en que una persona puede seleccionar cuatro productos de una lista de ocho.
def seleccionar_productos():
    return math.comb(8, 4)

# 17. ¿En cuántas formas pueden escogerse cuatro interruptores buenos y dos defectuosos de un lote que contiene 20 interruptores buenos y cinco defectuosos?
def seleccionar_interruptores():
    buenos = math.comb(20, 4)
    defectuosos = math.comb(5, 2)
    return buenos * defectuosos

# 18. Si se tienen tres canicas amarillas, dos azules y cuatro verdes, ¿de cuántas formas diferentes pueden acomodarse en una fila?
def acomodar_canicas():
    return math.factorial(9) // (math.factorial(3) * math.factorial(2) * math.factorial(4))

# 19. ¿Cuántas permutaciones habría con tres de los 10 títulos negociados cierto día en la bolsa de valores?
def permutaciones_titulos():
    return math.perm(10, 3)

# 20. Determine si cada una de las siguientes expresiones es verdadera o falsa:
def verificar_expresiones():
    a = math.factorial(9) == 9 * 8 * 7 * math.factorial(6)
    b = math.factorial(5) * math.factorial(4) == math.factorial(20)
    c = math.factorial(5) + math.factorial(5) == math.factorial(10)
    d = math.factorial(8) == math.factorial(9) // 9
    e = math.factorial(5) == math.factorial(10) // math.factorial(2)
    f = 3/4 + 1/4 == math.factorial(1)
    g = math.factorial(3) * math.factorial(2) == math.factorial(6)
    h = math.factorial(8) == math.factorial(6) * 56
    return a, b, c, d, e, f, g, h

# 21. El representante de un sindicato desea hablar con tres de los 10 trabajadores inmiscuidos en un procedimiento que es motivo de queja.
# a. Si es importante el orden de las entrevistas, ¿en cuántas formas puede planear las entrevistas el representante del sindicato?
def entrevistas_importa_orden():
    return math.perm(10, 3)

# b. Si no importa el orden de tres entrevistas, ¿de cuántas maneras puede planearlas el representante del sindicato?
def entrevistas_no_importa_orden():
    return math.comb(10, 3)

# 22. Calcula el número de formas que el alcalde de Mathprog puede elegir a 3 de 15 constructores para un ascenso.
def elegir_constructores():
    return math.comb(15, 3)

# 23. ¿En cuántas formas pueden alinearse seis reclutas para la ceremonia de izar la bandera, si tres reclutas son infantes de marina, uno del ejército, otro es de la naval y el último está en la fuerza aérea?
def alinear_reclutas():
    return math.factorial(6)

# 24. Un inversionista desea alimentar siete de las opciones de inversión de su portafolio vendiendo cuatro títulos y tres bonos. ¿En cuántas formas se puede vender estos valores, si entre los 25 títulos que hay en el portafolio 13 son acciones y el resto corresponde a bonos?
def vender_valores():
    acciones = math.comb(13, 4)
    bonos = math.comb(12, 3)
    return acciones * bonos

# 25. Una caja contiene una docena de focos eléctricos, que incluyen uno defectuoso. ¿En cuántas formas se pueden seleccionar dos focos, de modo que:
# a. No se incluya el foco defectuoso;
def seleccionar_focos_sin_defectuoso():
    return math.comb(11, 2)

# b. Se incluya el foco defectuoso.
def seleccionar_focos_con_defectuoso():
    return math.comb(11, 1)

# 26. Supóngase que, entre la docena de focos eléctricos del ejercicio anterior, hay dos unidades defectuosas. ¿De cuántas maneras pueden seleccionarse tres focos, de manera que
# a. No se incluya ninguno de los focos defectuosos;
def seleccionar_focos_sin_defectuosos():
    return math.comb(10, 3)

# b. Se incluya una de las unidades defectuosas;
def seleccionar_focos_un_defectuoso():
    return math.comb(10, 2) * math.comb(2, 1)

# c. Se incluyan ambos focos defectuosos.
def seleccionar_focos_dos_defectuosos():
    return math.comb(10, 1)

# 27. Si el productor de un noticiero tiene que escoger tres de siete noticias para transmitirlas al aire, ¿de cuántas maneras diferentes puede ordenar el noticiero?
def ordenar_noticias():
    return math.perm(7, 3)

# Ejecución de las funciones
print("12. Palabras con 4 consonantes y 3 vocales:", palabras_consonantes_vocales())
print("13. Permutaciones y combinaciones de alumnos:", permutaciones_combinaciones_alumnos())
print("14. Ordenar 3 libros escogiendo 2:", ordenar_libros())
print("15. Seleccionar 3 de 5 tiendas:", seleccionar_tiendas())
print("16. Seleccionar 4 productos de 8:", seleccionar_productos())
print("17. Seleccionar interruptores buenos y defectuosos:", seleccionar_interruptores())
print("18. Acomodar canicas:", acomodar_canicas())
print("19. Permutaciones de títulos:", permutaciones_titulos())
print("20. Verificar expresiones:", verificar_expresiones())
print("21a. Entrevistas con orden:", entrevistas_importa_orden())
print("21b. Entrevistas sin orden:", entrevistas_no_importa_orden())
print("22. Elegir constructores:", elegir_constructores())
print("23. Alinear reclutas:", alinear_reclutas())
print("24. Vender valores:", vender_valores())
print("25a. Seleccionar focos sin defectuoso:", seleccionar_focos_sin_defectuoso())
print("25b. Seleccionar focos con defectuoso:", seleccionar_focos_con_defectuoso())
print("26a. Seleccionar focos sin defectuosos:", seleccionar_focos_sin_defectuosos())
print("26b. Seleccionar focos con un defectuoso:", seleccionar_focos_un_defectuoso())
print("26c. Seleccionar focos con dos defectuosos:", seleccionar_focos_dos_defectuosos())
print("27. Ordenar noticias:", ordenar_noticias())

# 28. ¿De cuántas maneras diferentes puede seleccionarse un archivista cuatro expedientes de un gabinete que contiene 15?
def seleccionar_expedientes():
    return math.comb(15, 4)

# 29. ¿En cuántas formas se puede inscribir un estudiante en dos cursos de inteligencia artificial, dos de análisis de algoritmos y uno de computación?
def inscribir_cursos():
    ia = math.comb(6, 2)
    algoritmos = math.comb(4, 2)
    computacion = math.comb(3, 1)
    return ia * algoritmos * computacion

# 30. ¿En cuántas formas se puede asignar el desarrollo de tres personajes a tres programadores de videojuegos en una oficina, de manera que Fabio haga dos personajes, Sofía tres y Clara cinco?
def asignar_personajes():
    return math.factorial(10) // (math.factorial(2) * math.factorial(3) * math.factorial(5))

# 31. Si una prueba se compone de 12 preguntas de verdadero-falso,
# a. ¿de cuántas maneras diferentes un estudiante puede dar una respuesta para cada pregunta?
def respuestas_prueba():
    return 2 ** 12

# b. Si de antemano el profesor le dice que la primera pregunta es verdadera, ¿cuántas maneras tiene de contestar esta prueba?
def respuestas_prueba_con_primera_verdadera():
    return 2 ** 11

# 32. ¿De cuántas maneras ordenadas puede programar un director de televisión seis comerciales en los seis intermedios para comerciales durante la transmisión televisiva del primer tiempo del Super Bowl?
# a. Los comerciales son todos diferentes.
def comerciales_diferentes():
    return math.factorial(6)

# b. Dos de los comerciales son iguales.
def comerciales_dos_iguales():
    return math.factorial(6) // math.factorial(2)

# c. Si hay cuatro comerciales diferentes, uno de los cuales debe aparecer tres veces, mientras que cada uno de los otros debe aparecer una sola vez.
def comerciales_uno_tres_veces():
    return math.factorial(6) // math.factorial(3)

# 33. Una caja de 12 baterías recargables, contiene una defectuosa, ¿de cuántas maneras un inspector puede seleccionar tres de las baterías y:
# a. Obtener la defectuosa.
def seleccionar_baterias_con_defectuosa():
    return math.comb(11, 2)

# b. No obtener la defectuosa.
def seleccionar_baterias_sin_defectuosa():
    return math.comb(11, 3)

# 34. ¿De cuántas maneras pueden seleccionarse dos motores y dos conmutadores para un experimento de una antena de rastreo?
def seleccionar_motores_conmutadores():
    motores = math.comb(8, 2)
    conmutadores = math.comb(5, 2)
    return motores * conmutadores

# 35. La fábrica de automóviles Summer fabrica su popular modelo Sunshine en 10 colores, 5 líneas, 3 tipos de transmisión y 2 cilindrajes diferentes:
# a. ¿Cuántos tipos de vehículos se puede fabricar?
def tipos_vehiculos():
    return 10 * 5 * 3 * 2

# b. Si solamente se ofrecen colores primarios (rojo, amarillo y azul), ¿cuántos tipos se tendrán ahora?
def tipos_vehiculos_colores_primarios():
    return 3 * 5 * 3 * 2

# 36. ¿Cuántas placas de un automóvil se pueden hacer que contengan tres letras (sin incluir la Ñ) seguidas de tres dígitos y si se permita que haya repeticiones de letras y números? ¿Cuántas si no hay repeticiones?
def placas_con_repeticiones():
    return 26 ** 3 * 10 ** 3

def placas_sin_repeticiones():
    return math.perm(26, 3) * math.perm(10, 3)

# 37. ¿Cuántas cadenas de 16 bits comienzan y terminan con números 00?
def cadenas_16_bits():
    return 2 ** 14

# 38. ¿De cuántas maneras se puede seleccionar un presidente, vicepresidente, secretario y tesorero de un grupo de 12 estudiantes?
def seleccionar_directiva():
    return math.perm(12, 4)

# 39. Un coleccionista de libros antiguos desea ubicar sus 9 libros más preciados en una vitrina antirrobos, uno al lado del otro. 4 de los libros están escritos en griego y los 5 restantes en latín.
# a. ¿De cuántas formas se puede ubicar los libros en la vitrina?
def ubicar_libros():
    return math.factorial(9)

# b. ¿De cuántas formas si todos los libros en latín deben estar uno al lado del otro?
def ubicar_libros_latin_juntos():
    return math.factorial(5) * math.factorial(4) * math.comb(9, 5)

# c. Si desea alternar los libros (latín, griego, latín, griego, etc.), ¿de cuántas formas se puede ubicar ahora?
def ubicar_libros_alternados():
    return math.factorial(5) * math.factorial(4)

# Ejecución de las funciones
print("28. Seleccionar expedientes:", seleccionar_expedientes())
print("29. Inscribir cursos:", inscribir_cursos())
print("30. Asignar personajes:", asignar_personajes())
print("31a. Respuestas prueba:", respuestas_prueba())
print("31b. Respuestas prueba con primera verdadera:", respuestas_prueba_con_primera_verdadera())
print("32a. Comerciales diferentes:", comerciales_diferentes())
print("32b. Comerciales dos iguales:", comerciales_dos_iguales())
print("32c. Comerciales uno tres veces:", comerciales_uno_tres_veces())
print("33a. Seleccionar baterías con defectuosa:", seleccionar_baterias_con_defectuosa())
print("33b. Seleccionar baterías sin defectuosa:", seleccionar_baterias_sin_defectuosa())
print("34. Seleccionar motores y conmutadores:", seleccionar_motores_conmutadores())
print("35a. Tipos de vehículos:", tipos_vehiculos())
print("35b. Tipos de vehículos con colores primarios:", tipos_vehiculos_colores_primarios())
print("36a. Placas con repeticiones:", placas_con_repeticiones())
print("36b. Placas sin repeticiones:", placas_sin_repeticiones())
print("37. Cadenas de 16 bits:", cadenas_16_bits())
print("38. Seleccionar directiva:", seleccionar_directiva())
print("39a. Ubicar libros:", ubicar_libros())
print("39b. Ubicar libros latín juntos:", ubicar_libros_latin_juntos())
print("39c. Ubicar libros alternados:", ubicar_libros_alternados())
```
