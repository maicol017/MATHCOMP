```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial

# Datos de la tabla
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([-0.9, 0, 2, 4.5, 8.3, 13, 18])

# Ajuste del polinomio de segundo grado
p = Polynomial.fit(x, y, 2)

# Coeficientes del polinomio
coefs = p.convert().coef
print("Coeficientes del polinomio:", coefs)

# Calcular el valor ajustado
y_fit = p(x)

# Calcular el coeficiente de correlaci√≥n
correlation_matrix = np.corrcoef(y, y_fit)
correlation_coefficient = correlation_matrix[0, 1]
print("Coeficiente de correlaci√≥n (r):", correlation_coefficient)

# Graficar los datos y el polinomio ajustado
plt.scatter(x, y, label='Datos originales')
plt.plot(x, y_fit, label='Polinomio ajustado', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
